```python
#process heart rate data
def process_dataset(df):
    df = df.drop(df.index[:7])
    df = df.reset_index(drop=True)
    df = df[['Description and notes:', 'Unnamed: 1']]
    df.columns = ['time_offset', 'heartrate']
    df['time_seconds'] = pd.to_timedelta(df['time_offset']).dt.total_seconds()  # Convert to seconds
    df['heartrate'] = pd.to_numeric(df['heartrate'], errors='coerce')
    return df

# Process each dataset
hr_game1 = process_dataset(hr_game1)
hr_game2 = process_dataset(hr_game2)
hr_game3 = process_dataset(hr_game3)
hr_game4 = process_dataset(hr_game4)
hr_game5 = process_dataset(hr_game5)
hr_game6 = process_dataset(hr_game6)
```

```python
#plot heart rates
def plot_heart_rate(df, title):
    plt.figure(figsize=(10, 6))
    plt.plot(df['time_seconds'], df['heartrate'], marker='o')
    plt.title(title)
    plt.xlabel('Time Offset(seconds')
    plt.ylabel('Heart Rate')
    plt.grid(True)
    plt.show()

# Plot each game
plot_heart_rate(hr_game1, 'Heart Rate vs Time Offset - Game 1')
plot_heart_rate(hr_game2, 'Heart Rate vs Time Offset - Game 2')
plot_heart_rate(hr_game3, 'Heart Rate vs Time Offset - Game 3')
plot_heart_rate(hr_game4, 'Heart Rate vs Time Offset - Game 4')
plot_heart_rate(hr_game5, 'Heart Rate vs Time Offset - Game 5')
```

```python
def extract_hr(hr, df, game):
    hr = hr[['time_seconds', 'heartrate']]
    filtered_df = df[df['game'] == game]
    if filtered_df.empty:
        print(f"No data for game {game}")
        return df
    hits = filtered_df['heartrate_adj_time'].values
    print("hits for game", game, hits)

    processed_hits = 0  # To count the number of processed hits

    for hit in hits:
        # Define time intervals
        intervals = {
            'before3': (hit - 3, hit - 2),
            'before2': (hit - 2, hit - 1),
            'before1': (hit - 1, hit),
            'after1': (hit, hit + 1),
            'after2': (hit + 1, hit + 2),
            'after3': (hit + 2, hit + 3)
        }

        hit_processed = False  # To check if the current hit is processed

        # For each interval, calculate mean, min, and max heart rates
        for metric, (start, end) in intervals.items():
            values = hr.loc[(hr['time_seconds'] > start) & (hr['time_seconds'] <= end), 'heartrate'].values
            if len(values) > 0:
                mean_val = values.mean()
                min_val = values.min()
                max_val = values.max()
                hit_processed = True  # Mark that we have processed this hit for at least one interval
                print(f'{metric} interval for hit {hit}: {values}, mean: {mean_val}, min: {min_val}, max: {max_val}')
            else:
                mean_val = min_val = max_val = np.nan
                print(f'{metric} interval for hit {hit}: No data')

            # Store the calculated values in the corresponding columns
            df.loc[df['heartrate_adj_time'] == hit, f'{metric}_mean'] = mean_val
            df.loc[df['heartrate_adj_time'] == hit, f'{metric}_min'] = min_val
            df.loc[df['heartrate_adj_time'] == hit, f'{metric}_max'] = max_val

        if hit_processed:
            processed_hits += 1

    return df



# Initialize columns in the DataFrame for the new metrics
metrics = ['before3', 'before2', 'before1', 'after1', 'after2', 'after3']
for metric in metrics:
    df_final[f'{metric}_mean'] = np.nan
    df_final[f'{metric}_min'] = np.nan
    df_final[f'{metric}_max'] = np.nan

# Process the DataFrame
df_final = extract_hr(hr_game1, df_final, 1)
df_final = extract_hr(hr_game2, df_final, 2)
df_final = extract_hr(hr_game3, df_final, 3)
df_final = extract_hr(hr_game4, df_final, 4)
df_final = extract_hr(hr_game5, df_final, 5)
df_final = extract_hr(hr_game6, df_final, 6)

# Display the updated DataFrame
print("\nUpdated DataFrame:")
print(df_final)
```

```python
def extract_acc(df, acc, game):
    acc= acc[['Time (s)','Absolute acceleration (m/s^2)']]
    acc.rename(columns = {'Time (s)':'time'},inplace = True)
    # Initialize a dictionary with time (s) as keys and empty lists as values

    game_df = df[df['game'] == game]
    hits = game_df['accelerator_signal_hit_adj_time'].values

    acc_values = []
    acc_times = []
    time_diffs = []
    for hit in hits:

        before3_values = acc.loc[(acc['time'] >= hit - 2) & (acc['time'] <= hit), ['time','Absolute acceleration (m/s^2)']]
        if not before3_values.empty:
            max_acc_row = before3_values.loc[before3_values['Absolute acceleration (m/s^2)'].idxmax()]

            max_acc_value = max_acc_row['Absolute acceleration (m/s^2)']
            max_acc_time = max_acc_row['time']

            acc_values.append(max_acc_value)
            acc_times.append(max_acc_time)
            time_diffs.append(max_acc_time - hit)
        else:
            acc_values.append(np.nan)
            time_diffs.append(np.nan)
            acc_times.append(np.nan)


    df.loc[df_final['game'] == game, 'max_acc_value'] = acc_values
    df.loc[df_final['game'] == game, 'max_acc_time'] = acc_times
    df.loc[df_final['game'] == game, 'time_diff_max_acc'] = time_diffs

    return df


df_final = extract_acc(df_final, df_acc_1,1)
df_final = extract_acc(df_final, df_acc_2,2)
df_final = extract_acc(df_final, df_acc_3,3)
df_final = extract_acc(df_final, df_acc_4,4)
df_final = extract_acc(df_final, df_acc_5,5)
df_final = extract_acc(df_final, df_acc_5,6)
print(df_final)
```
